```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
x = np.linspace(-2,2)
for n in [1,2,3,4]:
    f = sp.hermite(n)
    y = f(x)
    plt.plot(x, y)
plt.grid()
```
